**Listamos as colunas que representam as contagens de crimes e criamos uma nova coluna que realiza a soma de todas essas colunas de crimes. Em seguida, somamos todas as ocorrências por ano e exibimos os resultados tanto em forma numérica quanto em gráfico para facilitar o entendimento.**

```
# Lista das colunas que representam contagens de crimes
colunas_de_crimes = df.columns[df.columns.get_loc('homicidio_doloso'):df.columns.get_loc('furto_de_veiculo')+1]

# Cria uma nova coluna que é a soma de todas as colunas de crimes
df['total_de_crimes'] = df[colunas_de_crimes].sum(axis=1)

# Soma todas as ocorrências de crimes por ano
crimes_por_ano = df.groupby('ano')['total_de_crimes'].sum()

# Exibe o número total de ocorrências por ano
print(crimes_por_ano)
```

```
import matplotlib.pyplot as plt

# Cria uma figura e um conjunto de subplots
fig, ax = plt.subplots()

# Cria um gráfico de linhas com os anos no eixo x e o número total de ocorrências no eixo y
ax.plot(crimes_por_ano.index, crimes_por_ano.values)

# Adiciona títulos e rótulos
ax.set_title('Número Total de Ocorrências por Ano')
ax.set_xlabel('Ano')
ax.set_ylabel('Número Total de Ocorrências')

# Mostra o gráfico
plt.show()
```

**Listamos as colunas que representam as contagens de crimes e criamos uma nova coluna que soma todas essas colunas. Em seguida, calculamos a soma de todas as ocorrências por região e ordenamos as regiões pela soma total de crimes. Exibimos as regiões em ordem decrescente pelo total de crimes. Com base nos resultados apresentados acima, podemos confirmar que o número total de ocorrências corresponde ao número total calculado anteriormente.**

```
colunas_de_crimes = df.columns[df.columns.get_loc('homicidio_doloso'):df.columns.get_loc('furto_de_veiculo')+1]
df['total_de_crimes'] = df[colunas_de_crimes].sum(axis=1)

crimes_por_regiao = df.groupby('regiao_ssp')['total_de_crimes'].sum()

crimes_por_regiao_ordenado = crimes_por_regiao.sort_values(ascending=False)

print(crimes_por_regiao_ordenado)

soma_total_de_crimes = df['total_de_crimes'].sum()

print(f"A soma total de ocorrências é: {soma_total_de_crimes:.0f}")


```

**Primeiramente, definimos uma lista contendo todas as regiões de interesse. Em seguida, implementamos um laço de repetição que percorre cada região na lista. Para cada região, filtramos os dados correspondentes e verificamos se há informações suficientes disponíveis. Caso uma região não possua dados suficientes, uma mensagem é impressa indicando a falta de dados para essa região específica. Para as regiões com dados suficientes, realizamos a soma das ocorrências de cada tipo de crime.**

```
# Lista das regiões que você mencionou
regioes = ['Capital', 'Grande São Paulo', 'Campinas', 'Ribeirao Preto', 'Piracicaba', 'Santos', 'Sorocaba', 'São Jose dos Campos', 'Bauru', 'São Jose do Rio Preto', 'Presidente Prudente', 'Araçatuba']

# Para cada região na lista de regiões
for regiao in regioes:
    # Filtra os dados para a região atual
    df_regiao = df[df['regiao_ssp'] == regiao]

    # Verifica se há dados suficientes para a região atual
    if not df_regiao.empty:
        # Soma as ocorrências de cada tipo de crime
        crimes_regiao = df_regiao[colunas_de_crimes].sum()

        # Cria uma figura e um conjunto de subplots
        fig, ax = plt.subplots()

        # Cria um gráfico de barras com os tipos de crimes no eixo x e o número de ocorrências no eixo y
        ax.bar(crimes_regiao.index, crimes_regiao.values)

        # Adiciona títulos e rótulos
        ax.set_title(f'Número de Ocorrências de Cada Crime em {regiao}')
        ax.set_xlabel('Tipo de Crime')
        ax.set_ylabel('Número de Ocorrências')

        # Rotaciona os rótulos do eixo x para melhor visualização
        plt.xticks(rotation=90)

        # Mostra o gráfico
        plt.show()
    else:
        print(f"Não há dados suficientes para a região: {regiao}")
```
